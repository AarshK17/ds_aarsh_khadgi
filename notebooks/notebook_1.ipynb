```python
data2['Timestamp IST'] = pd.to_datetime(
    data2['Timestamp IST'],
    dayfirst=True,
    format='mixed'
)
```

```python
data1['trade_date'] = data1['date'].dt.date
```

```python
data2['trade_date'] = data2['Timestamp IST'].dt.date
```

```python
merged_df = data2.merge(
    data1[['trade_date', 'classification']],
    on='trade_date',
    how='left'
)
```

```python
merged_df.rename(
    columns={'classification': 'market_sentiment'},
    inplace=True
)
```

```python
merged_df.groupby('market_sentiment')['Closed PnL'].mean().plot(kind='bar')
plt.title('Average Profit by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Average Closed PnL')
plt.show()
```

```python
merged_df['market_sentiment'].value_counts().plot(kind='bar')
plt.title('Trade Count by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Number of Trades')
plt.show()
```

```python
pd.crosstab(
    merged_df['Side'],
    merged_df['market_sentiment']
).plot(kind='bar', stacked=True)

plt.title('Buy vs Sell Trades by Market Sentiment')
plt.xlabel('Trade Side')
plt.ylabel('Number of Trades')
plt.show()
```

```python
plt.scatter(
    merged_df['Size USD'],
    merged_df['Closed PnL'],
    alpha=0.4
)
plt.title('Trade Size vs Profit')
plt.xlabel('Trade Size')
plt.ylabel('Closed PnL')
plt.show()
```

```python
plt.scatter(
    merged_df['Fee'],
    merged_df['Closed PnL'],
    alpha=0.4
)
plt.title('Fee vs Profit Relationship')
plt.xlabel('Fee')
plt.ylabel('Closed PnL')
plt.show()
```
